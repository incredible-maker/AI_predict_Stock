```python
import yfinance as yf
import pandas as pd

def save_financial_data(ticker):
    stock = yf.Ticker(ticker)
    
    # Get income statement for the current year
    income_stmt = stock.get_income_stmt()
    
    # Filter columns for the previous year
    current_year = income_stmt.columns[0].year
    prev_year_income_stmt = income_stmt.iloc[:, income_stmt.columns.year == current_year - 1]
    
    # Get balance sheet for the current year
    balance_sheet = stock.balance_sheet

        
    # Filter columns for the previous year
    prev_year_balance_sheet = balance_sheet.iloc[:, balance_sheet.columns.year == current_year - 1]
    

    # Get cash flow statement for the current year
    cash_flow_stmt = stock.cashflow

    # Filter columns for the previous year
    prev_year_cash_flow_stmt = cash_flow_stmt.iloc[:, cash_flow_stmt.columns.year == current_year - 1]

    # Save income statement as Excel file
    income_stmt_file = f"{ticker}_income_statement.xlsx"
    with pd.ExcelWriter(income_stmt_file) as writer:
        income_stmt.to_excel(writer, sheet_name='Current Year')
        prev_year_income_stmt.to_excel(writer, sheet_name='Previous Year')
    
    # Save balance sheet as Excel file
    balance_sheet_file = f"{ticker}_balance_sheet.xlsx"
    with pd.ExcelWriter(balance_sheet_file) as writer:
        balance_sheet.to_excel(writer, sheet_name='Current Year')
        prev_year_balance_sheet.to_excel(writer, sheet_name='Previous Year')


    # Save cash flow statement as Excel file
    cash_flow_stmt_file = f"{ticker}_cash_flow_statement.xlsx"
    with pd.ExcelWriter(cash_flow_stmt_file) as writer:
        cash_flow_stmt.to_excel(writer, sheet_name='Current Year')
        prev_year_cash_flow_stmt.to_excel(writer, sheet_name='Previous Year')
    
    print(f"Income statement saved as {income_stmt_file}")
    print(f"Balance sheet saved as {balance_sheet_file}")
    print(f"Cash flow statement saved as {cash_flow_stmt_file}")



    # Read the balance sheet and cash flow statement for the current year
    balance_sheet_file = f"{ticker}_balance_sheet.xlsx"
    cash_flow_stmt_file = f"{ticker}_cash_flow_statement.xlsx"
    df_balance_sheet_current = pd.read_excel(balance_sheet_file, sheet_name='Current Year', index_col=0)
    df_cash_flow_stmt_current = pd.read_excel(cash_flow_stmt_file, sheet_name='Current Year', index_col=0)

    # Calculate the ratios
    df_balance_sheet_current.loc['Debt to Equity Ratio'] = df_balance_sheet_current.loc['Total Debt'] / df_balance_sheet_current.loc['Stockholders Equity']
    df_balance_sheet_current.loc['Quick Ratio'] = (df_balance_sheet_current.loc['Cash And Cash Equivalents'] + df_balance_sheet_current.loc['Other Short Term Investments']) / df_balance_sheet_current.loc['Current Liabilities']
    df_balance_sheet_current.loc['Current Ratio'] = df_balance_sheet_current.loc['Current Assets'] / df_balance_sheet_current.loc['Current Liabilities']
    df_balance_sheet_current.loc['Revenue Growth Rate'] = (df_balance_sheet_current.loc['Total Capitalization'] - df_balance_sheet_current.shift(1).loc['Total Capitalization']) / df_balance_sheet_current.shift(1).loc['Total Capitalization']
    df_balance_sheet_current.loc['Cash Flow rate'] = df_cash_flow_stmt_current.loc['Operating Cash Flow'] / df_balance_sheet_current.loc['Total Debt']

    # Create a new dataframe with ratio calculation
    df_ratio = df_balance_sheet_current.loc[['Debt to Equity Ratio', 'Quick Ratio', 'Current Ratio', 'Revenue Growth Rate', 'Cash Flow rate']]

    # Save the ratios as a new Excel file
    ratios_file = f"{ticker}_financial_ratios.xlsx"
    df_ratio.to_excel(ratios_file)

    print(f"Financial ratios saved as {ratios_file}")

    
    
# Example usage
ticker_symbol = input("Enter ticker symbol: ")
save_financial_data(ticker_symbol)
```

```python
import yfinance as yf
import pandas as pd


data_directory = "/home/jacky/AIfinance/AI_predict_Stock/data"


def save_financial_data(ticker):
    stock = yf.Ticker(ticker)



    # Get income statement for the current year
    income_stmt = stock.get_income_stmt()

    # Filter columns for the previous year
    current_year = income_stmt.columns[0].year
    prev_year_income_stmt = income_stmt.iloc[:, income_stmt.columns.year == current_year - 1]

    # Get balance sheet for the current year
    balance_sheet = stock.balance_sheet

    # Filter columns for the previous year
    prev_year_balance_sheet = balance_sheet.iloc[:, balance_sheet.columns.year == current_year - 1]

    # Get cash flow statement for the current year
    cash_flow_stmt = stock.cashflow

    # Filter columns for the previous year
    prev_year_cash_flow_stmt = cash_flow_stmt.iloc[:, cash_flow_stmt.columns.year == current_year - 1]

    # Save income statement as Excel file
    income_stmt_file = f"{ticker}_income_statement.xlsx"
    with pd.ExcelWriter(income_stmt_file) as writer:
        income_stmt.to_excel(writer, sheet_name='Current Year')
        prev_year_income_stmt.to_excel(writer, sheet_name='Previous Year')

    # Save balance sheet as Excel file
    balance_sheet_file = f"{ticker}_balance_sheet.xlsx"
    with pd.ExcelWriter(balance_sheet_file) as writer:
        balance_sheet.to_excel(writer, sheet_name='Current Year')
        prev_year_balance_sheet.to_excel(writer, sheet_name='Previous Year')

    # Save cash flow statement as Excel file
    cash_flow_stmt_file = f"{ticker}_cash_flow_statement.xlsx"
    with pd.ExcelWriter(cash_flow_stmt_file) as writer:
        cash_flow_stmt.to_excel(writer, sheet_name='Current Year')
        prev_year_cash_flow_stmt.to_excel(writer, sheet_name='Previous Year')

    print(f"Income statement saved as {income_stmt_file}")
    print(f"Balance sheet saved as {balance_sheet_file}")
    print(f"Cash flow statement saved as {cash_flow_stmt_file}")


def calculate_ratios(ticker_list):
    all_ratios = {}  # A dictionary to store ratios for each company

    for ticker in ticker_list:
        save_financial_data(ticker)  # Save financial statements as Excel files

        # Read the balance sheet and cash flow statement for the current year
        balance_sheet_file = f"{ticker}_balance_sheet.xlsx"
        cash_flow_stmt_file = f"{ticker}_cash_flow_statement.xlsx"
        df_balance_sheet_current = pd.read_excel(balance_sheet_file, sheet_name='Current Year', index_col=0)
        df_cash_flow_stmt_current = pd.read_excel(cash_flow_stmt_file, sheet_name='Current Year', index_col=0)

        # ... (rest of the code for calculating ratios, as previously shown)
        # Calculate the ratios
        df_balance_sheet_current.loc['Debt to Equity Ratio'] = df_balance_sheet_current.loc['Total Debt'] / df_balance_sheet_current.loc['Stockholders Equity']
        df_balance_sheet_current.loc['Quick Ratio'] = (df_balance_sheet_current.loc['Cash And Cash Equivalents'] + df_balance_sheet_current.loc['Other Short Term Investments']) / df_balance_sheet_current.loc['Current Liabilities']
        df_balance_sheet_current.loc['Current Ratio'] = df_balance_sheet_current.loc['Current Assets'] / df_balance_sheet_current.loc['Current Liabilities']
        df_balance_sheet_current.loc['Revenue Growth Rate'] = (df_balance_sheet_current.loc['Total Capitalization'] - df_balance_sheet_current.shift(1).loc['Total Capitalization']) / df_balance_sheet_current.shift(1).loc['Total Capitalization']
        df_balance_sheet_current.loc['Cash Flow rate'] = df_cash_flow_stmt_current.loc['Operating Cash Flow'] / df_balance_sheet_current.loc['Total Debt']

        # Create a new dataframe with ratio calculation
        df_ratio = df_balance_sheet_current.loc[['Debt to Equity Ratio', 'Quick Ratio', 'Current Ratio', 'Revenue Growth Rate', 'Cash Flow rate']]

        all_ratios[ticker] = df_ratio  # Store the ratios for this company in the dictionary

    return all_ratios


def compare_companies(ticker_list):
    all_company_ratios = calculate_ratios(ticker_list)
    df_list = []

    for ticker, ratios in all_company_ratios.items():
        df = pd.DataFrame(ratios).T
        df['Company'] = ticker
        df_list.append(df)

    df_comparison = pd.concat(df_list, ignore_index=True)

    return df_comparison

# Example usage
ticker_symbols = ["AAPL", "MSFT", "GOOG", "ADBE"]

df_comparison = compare_companies(ticker_symbols)

#save it as a csv file
df_comparison.to_csv(f"{data_directory}/financial_ratios.csv", index=False)

# Print the DataFrame containing the financial ratios for the input companies
print(df_comparison)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the financial ratios data from the CSV file
df1 = pd.read_csv("/home/jacky/AIfinance/AI_predict_Stock/data/financial_ratios.csv")

# Calculate the average financial ratios for each company
average_ratios = df1.groupby('Company').mean()

# Transpose the DataFrame to have ratios as columns and companies as rows
average_ratios_transposed = average_ratios.T

# Plot the average financial ratios for each company
average_ratios_transposed.plot(kind='bar', figsize=(10, 6))
plt.xlabel('Financial Ratio')
plt.ylabel('Average Value')
plt.title('Average Financial Ratios for Each Company')
plt.legend(loc='upper right')
plt.show()
```
